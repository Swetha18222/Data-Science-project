```python
import pulp
from pulp import *
import pandas as pd
```

```python
df= pd.read_csv("/content/diet.csv")
df
```

```python
prob = LpProblem("Diet_Problem",LpMaximize)
prob
```

```python
df1=(df.iloc[:,0].reset_index())
df1
```

```python
food_list = df['Foods'].tolist()

price_dict = dict(zip(food_list, df['Price/Serving ($)']))
calories_dict = dict(zip(food_list, df['Calories']))
cholesterol_dict = dict(zip(food_list, df['Cholesterol (mg)']))
total_fat_dict = dict(zip(food_list, df['Total_Fat (g)']))
sodium_dict = dict(zip(food_list, df['Sodium (mg)']))
carbohydrates_dict = dict(zip(food_list, df['Carbohydrates (g)']))
fiber_dict = dict(zip(food_list, df['Dietary_Fiber (g)']))
protein_dict = dict(zip(food_list, df['Protein (g)']))
vit_a_dict = dict(zip(food_list, df['Vit_A (IU)']))
vit_c_dict = dict(zip(food_list, df['Vit_C (IU)']))
calcium_dict = dict(zip(food_list, df['Calcium (mg)']))
iron_dict = dict(zip(food_list, df['Iron (mg)']))

```

```python
iron_dict
```

```python
food_vars1=LpVariable.dicts('Foods',food_list,lowBound=0, cat="Continuous")
food_vars1
```

```python
prob+=lpSum(price_dict[food_item]* food_vars1[food_item] for food_item in food_list)
prob
```

```python
prob += lpSum(food_vars1[x]* calories_dict[x] for x in food_list)>=2.6, "CarbsMinimum"
prob += lpSum(food_vars1[x]* calories_dict[x] for x in food_list)<=358.2, "CarbsMaximum"
prob
```

```python
prob += lpSum(food_vars1[x]* calories_dict[x] for x in food_list)>=2.6, "CarbsMinimum"
prob += lpSum(food_vars1[x]* calories_dict[x] for x in food_list)<=358.2, "CarbsMaximum"

prob += lpSum(food_vars1[x]* cholesterol_dict[x] for x in food_list)>=0.0, "CholestrolMinimum"
prob += lpSum(food_vars1[x]* cholesterol_dict[x] for x in food_list)<=211.2, "CholestrolMaximum"

prob += lpSum(food_vars1[x]* vit_a_dict[x] for x in food_list)>=0.0, "VitAMinimum"
prob += lpSum(food_vars1[x]* vit_a_dict[x] for x in food_list)<=5867.4, "VitAMaximum"

prob += lpSum(food_vars1[x]* calcium_dict[x] for x in food_list)>=3.3, "CaMinimum"
prob += lpSum(food_vars1[x]* calcium_dict[x] for x in food_list)<=159.0, "CaMaximum"
```

```python
prob.solve()
```

```python
prob.solver
```

```python
LpStatus[prob.status]
```

```python
from pulp import LpProblem, LpVariable, lpSum, LpMaximize
for var in prob.variables():
  print(f'variable name: {var.name},variable value: {var.value()}\n')

print('\n')
print('*'*100)
print('\n')

for name, con in prob.constraints.items():
  print(f'constraints name: {name},constraints value: {con.value()}')

print('*'*100)
print('\n')

print(f'Objective Value is: {round(prob.objective.value(),2)}')
```

The optimal value of the diet is 4.58 which include all the food items.
