```python
import pandas as pd
d= pd.read_csv("/content/mood_data.txt",sep=";",header =None,names = ["Text","Emotions"])
d
```

```python
import nltk
from nltk import word_tokenize
import string
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('stopwords')
```

```python
def clean_text(mood):
    mood = word_tokenize(mood)
    mood= " ".join(mood)
    mood = [char for char in mood if char not in string.punctuation]
    mood = "".join(mood)
    mood = [word for word in mood.split() if mood.lower() not in stopwords.words('english')]
    return " ".join(mood)


```

```python
d["clean_text"] = d["Text"].apply(clean_text)
```

```python
d
```

```python
import re
```

```python
features = d['clean_text']
processed_features = []

for sentence in range(0, len(features)):

    processed_feature = re.sub(r'\W', ' ', str(features[sentence]))


    processed_feature= re.sub(r'\s+[a-zA-Z]\s+', ' ', processed_feature)


    processed_feature = re.sub(r'\^[a-zA-Z]\s+', ' ', processed_feature)


    processed_feature = re.sub(r'\s+', ' ', processed_feature, flags=re.I)


    processed_feature = re.sub(r'^b\s+', '', processed_feature)


    processed_feature = processed_feature.lower()

    processed_features.append(processed_feature)
```

```python
d['processed_text'] = processed_features
```

```python
d
```

```python
data = {
    'Emotion': d["Emotions"],
    'Processed_Text': d["processed_text"]}
```

```python
df=pd.DataFrame(data)
df
```

```python
from nltk.tokenize import TweetTokenizer
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
def tokenize(text):
    tk = TweetTokenizer()
    return tk.tokenize(text)

vectorizer = CountVectorizer(analyzer = 'word',tokenizer = tokenize,lowercase = True,ngram_range=(1, 1))
```

```python
count= vectorizer.fit_transform(df['Processed_Text'])
```

```python
x = df["Processed_Text"].values
y = df["Emotion"].values
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=1000)
X_train_idf = vectorizer.fit_transform(x_train)
X_test_idf = vectorizer.transform(x_test)
```

```python

df_idf = pd.DataFrame(vectorizer.idf_, index=vectorizer.get_feature_names_out(),columns=["idf_weights"])

df_idf.sort_values(by=['idf_weights'],ascending = False).head()
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
```

```python
m = MultinomialNB()
m.fit(X_train_idf, y_train)
```

```python
p = m.predict(X_test_idf)
acc = accuracy_score(y_test, p)
print('Multinomial Naive Bayes', acc)
```

```python
results = pd.DataFrame([['Multinomial Naive Bayes', acc]],columns = ['Model', 'Accuracy'])

print(results)
```

```python
from sklearn.ensemble import RandomForestClassifier
c = RandomForestClassifier()
c.fit(X_train_idf, y_train)
```

```python
y_pred = c.predict(X_test_idf)
c = accuracy_score(y_test, y_pred)

model_results = pd.DataFrame([['RandomForest(Gini)', c]],columns = ['Model', 'Accuracy'])
results = results.append(model_results, ignore_index = True)
print(results)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(criterion='entropy')
clf.fit(X_train_idf, y_train)
```

```python
y = clf.predict(X_test_idf)


a = accuracy_score(y_test, y)
model_results = pd.DataFrame([['RandomForest(Entropy)', a]],columns = ['Model', 'Accuracy'])
results = results.append(model_results, ignore_index = True)
print(results)
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train_idf, y_train)
```

```python
b = svc.predict(X_test_idf)

# Calculate accuracy
acc = accuracy_score(y_test, b)
model_results = pd.DataFrame([['SVM by SVM', acc]],columns = ['Model', 'Accuracy'])
results = results.append(model_results, ignore_index = True)
print(results)
```
